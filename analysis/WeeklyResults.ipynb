Download the corresponding outputs



```python
import numpy as np
from numpy import array
import matplotlib.pyplot as plt
import os

filepath = f"data/weeklyAddition/week{weekno}SubmissionProcessed"
os.makedirs(filepath, exist_ok=True)

# Construct dynamic URL
url = baseUrl + filepath + "/outputs.txt"

# Download using wget
!wget -O $filepath/outputs.txt $url

all_weeks_array = []
current_line = ""

with open(filepath + "/outputs.txt", 'r') as f:
    for line in f:
        stripped = line.strip()
        if not stripped:
            continue
        current_line += stripped
        if stripped.endswith("]"):
            all_weeks_array.append(eval(current_line))
            current_line = ""

all_weeks_array = np.array(all_weeks_array)


print(all_weeks_array.shape)


```

This is to read the Weekly processed Submissions outputs.
As the intent of all the functions is to be maximised, sucess can be inferred by the output increasing or descreasing. Ideally a higher output is wanted to bring progress.

we see the results plotted per function

```python

final   = all_weeks_array[-1, :]  # Values for first week
initial = all_weeks_array.shape[1] # Values for last week

# Compare final vs initial
improvement = final > initial   # True if value increased
decline     = final < initial   # True if value decreased
no_change   = final == initial  # True if unchanged
action = ''


for i in range(initial):
    if improvement[i]:
        action = "improved +++"

    elif decline[i]:
        action ='decreased --'

    else:
        action = "stayed the same <->"

    functionOverWeeks = f"Function {i+1} {action} over {weekno} week{'s' if weekno > 1 else ''}."
    print(functionOverWeeks)
```

```python
print(all_weeks_array.shape[0])
weeks = np.arange(1, all_weeks_array.shape[0] + 1)

for i in range(all_weeks_array.shape[1]):
    color = 'green' if improvement[i] > 0 else 'red'
    plt.plot(weeks, all_weeks_array[:, i], marker='o', color=color, label=f'Funcion {i+1}')

plt.xlabel('Week')
plt.ylabel('Output Value')
plt.title(f"Progress per Function outputs over the {weekno} weeks (green=improved, red=declined)")
plt.legend()
plt.grid(True)
plt.show()
```

We see the increases and decreases per week and can see what functions need to have the current approach revaluated

```python

num_weeks = all_weeks_array.shape[0]
num_functions = all_weeks_array.shape[0]
function_names = {
    0: "2D Contamination",
    1: "2D Noisy Log-Likelihood",
    2: "3D Drug Combination",
    3: "4D Warehouse Placement",
    4: "4D Chemical Yield",
    5: "5D Cake Recipe",
    6: "6D ML Hyperparameters",
    7: "8D ML Hyperparameters"
}
num_functions = len(function_names)
weeks = np.arange(1, num_weeks + 1)


weekly_changes = np.diff(all_weeks_array, axis=0)
overall_change = all_weeks_array[-1] - all_weeks_array[0]
last_week_change = all_weeks_array[-1] - all_weeks_array[-2]

biggest_drop = np.argmin(last_week_change)
biggest_rise = np.argmax(last_week_change)

fraction_improved = np.mean(weekly_changes > 0, axis=0)
avg_weekly_change = np.mean(weekly_changes, axis=0)
combined_score = fraction_improved * avg_weekly_change

worst_funcs = np.argsort(combined_score)[:4]
best_funcs = np.argsort(combined_score)[-4:]


improvements = all_weeks_array[-1] - all_weeks_array[0]
print("Overall improvements per function:")
for i, imp in enumerate(improvements):
    print(f" Function {i+1} {function_names.get(i, f'Function {i+1}')}: {imp:.4f}")

print("\nFunctions needing immediate attention (worst improvement):")
for i in worst_funcs:
    print(f" (Function {i+1}) {function_names.get(i, f'Function {i+1}')} ")

print("\nFunctions improving the most (best improvement):")
for i in best_funcs:
    print(f" (Function {i+1}) {function_names.get(i, f'Function {i+1}')}")


# Plotting
for func_idx in range(num_functions):
    y_values = all_weeks_array[:, func_idx]

    # Determine line color
    if last_week_change[func_idx] < 0:
        line_color = 'red'
    elif overall_change[func_idx] > 0:
        line_color = 'green'
    else:
        line_color = 'gray'

    plt.figure()
    plt.plot(weeks, y_values, color=line_color, label=f'Function {func_idx+1}')

    # Max and min indices
    max_idx = np.argmax(y_values)
    min_idx = np.argmin(y_values)

    # Plot and annotate all points
    for i, val in enumerate(y_values):
        # Color logic for scatter points
        if i == max_idx:
            point_color = 'blue'
            point_label = 'Max Output'
        elif i == min_idx:
            point_color = 'orange'
            point_label = 'Min Output'
        else:
            point_color = 'magenta'
            point_label = None  # avoid duplicate legend entries

        plt.scatter(weeks[i], val, color=point_color, s=50, label=point_label)
        plt.annotate(f"{val:.8f}", (weeks[i], val), xytext=(5,5), textcoords="offset points", fontsize=8)

    # Dynamic title
    if func_idx == biggest_drop:
        plt.title(f"(Function {func_idx+1}) Biggest Weekly Drop: {function_names.get(func_idx)}")
    elif func_idx == biggest_rise:
        plt.title(f"(Function {func_idx+1}) Biggest Weekly Improvement: {function_names.get(func_idx)}")
    else:
        plt.title(f"(Function {func_idx+1}) {function_names.get(func_idx)} Progress Over {num_weeks} Weeks")

    plt.xlabel("Week")
    plt.ylabel("Score")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

Show the week to week changes per Function

```python

weekly_changes = np.diff(all_weeks_array, axis=0)


func_names = [f"Function {i+1} {function_names.get(i)} " for i in range(all_weeks_array.shape[1])]

for week_idx, changes in enumerate(weekly_changes, start=1):

    changes_print = [float(round(x, 4)) for x in changes]

    declines = [(func_names[i], changes_print[i]) for i in range(len(changes_print)) if changes_print[i] < 0]
    improvements = [(func_names[i], changes_print[i]) for i in range(len(changes_print)) if changes_print[i] > 0]
    unchanged = [(func_names[i], changes_print[i]) for i in range(len(changes_print)) if changes_print[i] == 0]


    declines.sort(key=lambda x: x[1])  # most negative first
    improvements.sort(key=lambda x: x[1], reverse=True)  # largest positive first


    highlight = " <- LAST WEEK" if week_idx == weekly_changes.shape[0] else ""
    print(f"Week {week_idx} -> Week {week_idx+1}{highlight}:")
    print("  Declined:", declines)
    print("  Improved:", improvements)
    print("  Unchanged:", unchanged, "\n")
```

```python



weekly_change = np.diff(all_weeks_array, axis=0)  # week i -> week i+1

num_points = all_weeks_array.shape[1]

for j in range(num_points):

    decline_values = weekly_change[:, j]
    biggest_decline = np.min(decline_values)

    if biggest_decline < 0:
        week_index = np.argmin(decline_values)  # index of the biggest decline
        print(f"Function {j+1} {function_names.get(j)} : Biggest decline = {biggest_decline:.3f} during Week {week_index+1} -> Week {week_index+2}")
    else:
        print(f"Function {j+1} {function_names.get(j)} : No decline occurred")
```

```python
from IPython.display import display, HTML
import numpy as np

weekly_change = np.diff(all_weeks_array, axis=0)  # week i -> week i+1
num_points = all_weeks_array.shape[1]

html_output = ""

for j in range(num_points):
    increase_values = weekly_change[:, j]
    biggest_increase = np.max(increase_values)

    if biggest_increase > 0:
        week_index = np.argmax(increase_values)
        html_output += f"<p style='font-family:Courier; font-size:14px;'>"
        html_output += f"<b>Function {j+1} {function_names.get(j)} :</b> Biggest increase = {biggest_increase:.3f} "
        html_output += f"during Week {week_index+1} → Week {week_index+2}</p>"
    else:
        html_output += f"<p style='font-family:Courier; font-size:14px;'>Function {j+1}: No increase occurred</p>"

display(HTML(html_output))


```
